```python
#Nous allons rechercher quel est le lien entre la puissance (horsepower) et la taille du moteur (enginesize).

plt.plot(raw_cars['enginesize'], raw_cars['horsepower'], 'ro')
```

```python
#Faisons la distinction entre les moteurs turbo et les autres
condition = raw_cars['aspiration']=='turbo'
turbo_cars = raw_cars[condition]
condition = raw_cars['aspiration']=='std'
atmo_cars = raw_cars[condition]
atmo_cars = atmo_cars[atmo_cars['horsepower']]
atmo_cars
```

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
ax = ax.set(xlabel='engine size', ylabel='horsepower')

gas_cars = raw_cars[raw_cars['fueltype'] == 'gas']
other_cars = raw_cars[raw_cars['fueltype'] != 'gas']

plt.scatter(gas_cars['enginesize'], gas_cars['horsepower'], marker='p', c='green')
plt.scatter(other_cars['enginesize'], other_cars['horsepower'], marker='v', c='yellow')
plt.show()
```

```python
# retirons les données abérante
turbo_cars = turbo_cars[turbo_cars['horsepower'] < 200]
atmo_cars = atmo_cars[atmo_cars['horsepower'] < 200]

#Faisons un régression linéaire pour les moteurs atmosphériques.
v0 = np.ones(atmo_cars.shape[0])
v1 = atmo_cars['enginesize'].values
X = np.matrix([v0,v1]).T
y = np.matrix(atmo_cars['horsepower']).T

theta_atmo = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

print(theta_atmo)


v0 = np.ones(turbo_cars.shape[0])
v1 = turbo_cars['enginesize'].values
X = np.matrix([v0,v1]).T
y = np.matrix(turbo_cars['horsepower']).T

theta_turbo = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
print(theta_turbo)
```

```python
plt.scatter(turbo_cars['enginesize'], turbo_cars['horsepower'], marker='p')
plt.scatter(atmo_cars['enginesize'], atmo_cars['horsepower'], marker='v')

plt.plot([0,350],[theta_atmo.item(0), theta_atmo.item(0)+350 * theta_atmo.item(1)], linestyle='--', c='orange')
plt.plot([0,350],[theta_turbo.item(0), theta_turbo.item(0)+350 * theta_turbo.item(1)], linestyle='--', c='blue')
plt.show()
```

```python
#reitérons l'expérience avec les différents type de carburants. 
gas_cars = raw_cars[raw_cars['fueltype'] == 'gas']
other_cars = raw_cars[raw_cars['fueltype'] != 'gas']


v0 = np.ones(gas_cars.shape[0])
v1 = gas_cars['enginesize'].values
X = np.matrix([v0,v1]).T
y = np.matrix(gas_cars['horsepower']).T
theta_gas = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
print(theta_gas)

v0 = np.ones(other_cars.shape[0])
v1 = other_cars['enginesize'].values
X = np.matrix([v0,v1]).T
y = np.matrix(other_cars['horsepower']).T
theta_other = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
print(theta_other)

plt.scatter(gas_cars['enginesize'], gas_cars['horsepower'], marker='p', c='green')
plt.scatter(other_cars['enginesize'], other_cars['horsepower'], marker='v', c='yellow')
plt.plot([0,350],[theta_gas.item(0), theta_gas.item(0)+350 * theta_gas.item(1)], linestyle='--', c='green')
plt.plot([0,350],[theta_other.item(0), theta_other.item(0)+350 * theta_other.item(1)], linestyle='--', c='yellow')
plt.show()
```

```python
#Affichons nos 4 droites. 
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
ax = ax.set(xlabel='engine size', ylabel='horsepower')

plt.plot([0,350],[theta_atmo.item(0), theta_atmo.item(0)+350 * theta_atmo.item(1)], linestyle='--', c='orange')
plt.plot([0,350],[theta_turbo.item(0), theta_turbo.item(0)+350 * theta_turbo.item(1)], linestyle='--', c='blue')
plt.plot([0,350],[theta_gas.item(0), theta_gas.item(0)+350 * theta_gas.item(1)], linestyle='--', c='green')
plt.plot([0,350],[theta_other.item(0), theta_other.item(0)+350 * theta_other.item(1)], linestyle='--', c='yellow')
plt.show()
```
